```python
table_list=[]
for url in urls:
    response=requests.get(url)
    content=response.content
    soup=BeautifulSoup(content,'html.parser')
    all_tables=soup.find_all("table")
    req_table=all_tables[2]
    table_data=req_table.find_all("td")
    start=0
    for i in range(50):
        Player,	Span,	Mat,	RunsDescending,	HS,	Bat_Av,	hund,Wkts,BBI,Bowl_Av,fives,Ct,St,Avg_Diff=table_data[start:start+14]
        table_list.append([Player.text,	Span.text,	Mat.text,	RunsDescending.text,HS.text,Bat_Av.text,hund.text,Wkts.text,BBI.text,Bowl_Av.text,fives.text,Ct.text,St.text,Avg_Diff.text])
        start+=15
print(*table_list,sep='\n')
print(len(table_list))
```

```python
#finding column names
#find columns name from "th" tag
col=[]
for i in req_table.find_all('th'):
    col.append(i.text)
print(col)

data=pd.DataFrame(table_list,columns=col[:-1])
data.name='ODI matches played in India'
print(data.name)
data.head()
```

```python
#donwloading data into .csv file
data.to_csv('ODI_data.csv')
```
